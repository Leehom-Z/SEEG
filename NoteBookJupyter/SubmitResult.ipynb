# 2.相关总体结果的汇总(需要汇报的信息)

```python
import numpy as np
import os 
import pandas as pd 
from tqdm.notebook import tqdm
import sys
sys.path.append("../")
from util.util_file import time_add

def tuple_str_time(time_tuple):
    time_str = [str(x) for x in time_tuple]
    time_result = ":".join(time_str)
    return time_result

BDP_channel_list_path = "../data/data_slice/channels_info/BDP_seq.csv"
fi_id_nfi_dict_path = "./data/fi_nfi_id.pkl"
feature_id_time_spatial_path = "./data/feature_id_time_spatial.pkl"
Best_label_cluster_path = "./processing/Best_label_cluster.pkl"
BDP_val_prediction_path = "./data/BDP_val_prediction.pkl"
feature_id_time_dict_path = "./data/feature_id_time_dict.pkl"

BDP_channel = pd.read_csv(BDP_channel_list_path, sep=',')
IDs = BDP_channel['ID']
Chan_names = BDP_channel['chan_name']
channel_id_names = dict(zip(IDs, Chan_names))



fi_id_nfi_dict = np.load(fi_id_nfi_dict_path, allow_pickle=True)
Best_label_cluster  = np.load(Best_label_cluster_path, allow_pickle=True)
feature_id_time_spatial = np.load(feature_id_time_spatial_path, allow_pickle=True)
BDP_val_prediction = np.load(BDP_val_prediction_path, allow_pickle=True)
feature_id_time_dict = np.load(feature_id_time_dict_path, allow_pickle=True)

print(Best_label_cluster)
print(feature_id_time_spatial)
print(feature_id_time_dict)

Global_id = []
Cluster_id = []
Absolute_time_start = []
Absolute_time_end = []
Relative_time_start = []
Relative_time_end = []
Hot_spot_time = []
Hot_spot_time_sec = []  
Relative_time_sec_start = []
Relative_time_sec_end = []
File_name = []
Channel_name = []
for cluster_id, f_id_list in tqdm(Best_label_cluster.items()):
    for f_id in f_id_list:
        Global_id.append(f_id)
        Cluster_id.append(cluster_id)
        spot_time = feature_id_time_spatial[f_id]["time"]
        spot_time /= 100
        ab_time_ = feature_id_time_dict[f_id]["Absolute time"]
        ab_time_end = time_add(ab_time_[0], ab_time_[1], ab_time_[2], 2)
        Absolute_time_start.append(tuple_str_time(ab_time_))
        Absolute_time_end.append(tuple_str_time(ab_time_end))
        re_time_ = feature_id_time_dict[f_id]["Relative time"]
        re_spot_time = time_add(re_time_[0], re_time_[1], re_time_[2], spot_time)
        Hot_spot_time.append(tuple_str_time(re_spot_time))
        re_time_end = time_add(re_time_[0], re_time_[1], re_time_[2], 2)
        Relative_time_start.append(tuple_str_time(re_time_))
        Relative_time_end.append(tuple_str_time(re_time_end))
        re_time_sec = re_time_[0]*3600+re_time_[1]*60+re_time_[2]
        re_spot_time_sec = re_time_[0]*3600+re_time_[1]*60+spot_time
        Hot_spot_time_sec.append(re_spot_time_sec)
        Relative_time_sec_start.append(re_time_sec)
        Relative_time_sec_end.append(re_time_sec+2)
        _f_name = feature_id_time_dict[f_id]["File name"]
        File_name.append(_f_name)
        channel_number = feature_id_time_spatial[f_id]["spatial"]
        
        _chan_name = channel_id_names[channel_number]
        Channel_name.append(_chan_name)

result = {"Feature ID":Global_id,
          "Absolute start time":Absolute_time_start,
          "Absolute end time":Absolute_time_end, 
          "Relative start time":Relative_time_start,
          "Relative end time":Relative_time_end,
          "Hot spot": Hot_spot_time, 
          "Relative start time [sec]":Relative_time_sec_start,
          "Relative end time [sec]":Relative_time_sec_end, 
          "Hot spot [sec]":Hot_spot_time_sec,
          "Channel name":Channel_name,
          "Data Source":File_name,
          "Cluster ID":Cluster_id
          
          }
result_csv = pd.DataFrame(result)
result_csv.to_csv("./result/result.csv")
        
        
    

```
